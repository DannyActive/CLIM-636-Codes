```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
# Load data
filename = 'national_tier1_caps_09feb2024_0.xlsx'
data = pd.read_excel(filename, sheet_name='PM10Primary', header=0, index_col=0)
data.index.name=None
data
```

```python
# subset relevant data
# data = data.loc[4:18, :]
data = data.loc['Source Category': 'MISCELLANEOUS'] #select the relevant part rows of data
data
```

```python
data.columns = data.iloc[0] #replace the header with the first row of data
data = data.drop(data.index[0]) #drop the first row
data
```

```python
# Change names of columns
# data.columns = data.iloc[0]
# data = data.drop(data.index[0])
# data
```

```python
data.fillna(0, inplace=True) #fill replace NAN values with 0
data
```

```python
data = data.T #Traspose the data to have the years as the index
data
```

```python
# data.columns = data.iloc[0]
# data = data.drop(data.index[0])
# data
```

```python
# change the index(years) from float to interger
data.index = data.index.astype(int)
data.head()
```

```python
data.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Source Emissions of PM$_{10}$ in the US', fontsize=14)
plt.xlabel('Year', fontsize= 14)
plt.ylabel('Emissions (kt)', fontsize = 14)
plt.legend(loc = 0, ncols=3, fontsize=7)
plt.savefig('PM10.png', dpi=1200)
```
